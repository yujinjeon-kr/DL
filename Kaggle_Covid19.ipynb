```python
import pandas as pd

# csv 데이터 변수에 저장
confirmed_df = pd.read_csv('time_series_covid_19_confirmed.csv')
deaths_df = pd.read_csv('time_series_covid_19_deaths.csv')
recovered_df = pd.read_csv('time_series_covid_19_recovered.csv')
```

```python
# 확진자 수 데이터 중 대한민국만 추출, 대한민국에 해당하는 인덱스 번호 알아내기.-> 160
idx_korea = (confirmed_df['Country/Region']=='Korea, South')
idx_korea
```

```python
# 국내 확진자 추이
confirmed_korea_df = confirmed_df[idx_korea] # 알아낸 인덱스의 값만 변수 confirmed_korea_df에 저장
confirmed_korea_df
```

```python
import matplotlib.pyplot as plt
import numpy as np

confirmed_korea_np = np.array(confirmed_korea_df) # numpy 배열로 변환
confirmed_korea_np
```

```python
confirmed_korea_np.shape
```

```python
confirmed_korea_np = confirmed_korea_np[:,4:]
confirmed_korea_np
```

```python
confirmed_korea_np.shape
```

```python
# 2차원의 배열(1xn) 을 인덱싱을 사용하여 그래프로 올바르게 출력
plt.plot(confirmed_korea_np[0,:])      # 확진자 그래프 그리기

# 가로축 눈금 표시
label_x = confirmed_korea_df.columns.values[4:]  # DataFrame에서 인덱스 가져오기
sel_ids = range(0, len(label_x),100)            # 눈금을 표시할 위치
plt.xticks(sel_ids,labels=label_x[sel_ids])      # 눈금 표시
```

```python
# 누적 사망자 수 가져오기
idx_korea = (deaths_df['Country/Region']=='Korea, South')
deaths_korea_df = deaths_df[idx_korea]
deaths_korea_np = np.array(deaths_korea_df)
deaths_korea_np = deaths_korea_np[:,4:]
```

```python
# 누적 회복자 수 가져오기
idx_korea = (recovered_df['Country/Region']=='Korea, South')
recovered_korea_df = recovered_df[idx_korea]
recovered_korea_np = np.array(recovered_korea_df)
recovered_korea_np = recovered_korea_np[:,4:]
```

```python
plt.figure(figsize=(10,7))
plt.plot(confirmed_korea_np[0,:]) # 확진자 그래프 그리기
plt.plot(deaths_korea_np[0,:])    # 사망자 그래프 그리기
plt.plot(recovered_korea_np[0,:]) # 회복자 그래프 그리기


label_x = confirmed_korea_df.columns.values[4:] # DataFrame에서 인덱스 가져오기
sel_ids = range(0,len(label_x), 100) # 눈금을 표시할 위치 선택: 0부터 493(494-1)까지 100간격으로 -> 인덱스 0, 100, 200, 300, 400

plt.xticks(sel_ids, labels=label_x[sel_ids], fontsize=15) # x축 눈금 표시
plt.yticks(fontsize=15)                                   # y축 눈금 표시
plt.legend(['Confirmed', 'Death', 'Recovered'], fontsize=15)
```

```python
# 국내 확진자 대비 사망자 비율 추이
plt.figure(figsize=(10,7))
recovered2confirmed_ratio = deaths_korea_np[0,:]/ confirmed_korea_np[0,:]

plt.plot(recovered2confirmed_ratio)  # 확진자 대비 사망자 비율 그래프 그리기

label_x = confirmed_korea_df.columns.values[4:] # DataFrame에서 인덱스(날짜) 가져오기

sel_ids = range(0,len(label_x),100)
plt.xticks(sel_ids, labels=label_x[sel_ids], fontsize=15)
plt.yticks(fontsize=15)
```

```python
# 사망자 비율이 높게 나타난 일자와 값 찾기 : numpy의 통계함수 argmax 이용

idx_max = np.argmax(recovered2confirmed_ratio)

print(f'사망자/확진자 비율이 최대인 일자는: {label_x[idx_max]} 이며')
print(f'해당 일자의 값은 {recovered2confirmed_ratio[idx_max]:.3f}입니다.')
```
